# Instagram Scraping

Instagram caption generator.
- Scraping Instagram profiles for captions and images. 

```python
import instaloader
from instaloader import Instaloader, Profile

from itertools import islice
from math import ceil
```

```python
PROFILE = 'jukkarisikko'        # profile to download from
X_percentage = 100    # percentage of posts that should be downloaded

L = Instaloader(download_videos=False, download_video_thumbnails=True, download_comments=False, download_geotags=False, save_metadata=False)

profile = Profile.from_username(L.context, PROFILE)
posts_sorted_by_likes = (profile.get_posts())

for post in islice(posts_sorted_by_likes, ceil(profile.mediacount * X_percentage / 100)):
    L.download_post(post, PROFILE)
```
