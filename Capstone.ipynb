#### Non-Stationary Time Series
Observations from a non-stationary time series show seasonal effects, trends, and other structures that depend on the time index.

Summary statistics like the mean and variance do change over time, providing a drift in the concepts a model may try to capture.

Classical time series analysis and forecasting methods are concerned with making non-stationary time series data stationary by identifying and removing trends and removing seasonal effects.

```python
tbond = pd.read_csv('USTBOND_TRUSD_Candlestick_48_M_ASK_csv.csv')
```

```python
tbond = pd.DataFrame(tbond)
tbond.head()
```

```python
tbond.plot('Gmt time',['Open'], figsize=(15,10), rot=90)
plt.pyplot.show()
```

```python
def testing_stationarity(time_series):
    
    rolmean = time_series.rolling(window=5).mean()
    rolstd = time_series.rolling(window=5).std()
    
    fig = plt.figure(figsize=(15,10))
    ax = fig.
    origin = plt.plot(time_series, color='blue', label='Original Data')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='green', label='Rolling Std')
    plt.show()
    
    print('Dickey Fuller Test Results')
    dftest = adfuller(time_series)
    dfoutput = pd.Series(dftest, index=('T-Statistic', 'P-Value', 'Lags', 'Number obv', 'Critical Values'))
    print(dfoutput)
```

```python
testing_stationarity(tbond)
```
